```
print("Ejercicio_4") 

print("Este programa es una pequeña calculadora en la cual puede")
print("elegir que operaciones matematicas se pueden hacer")
print("Para SUMAR escriba A.")
print("Para RESTAR escriba B.")
print("Para MULTIPLICAR escriba C.")
print("Para DIVIDIR escriba D.")

opcion = str(input("¿Que desea hacer? "))
if opcion == "A" or opcion=="a" :
  print("Has elegido la opción de SUMAR, por lo cual debes ingresar dos cifras")
  a=float(input("Ingresa la primera cifra: "))
  b=float(input("Ingresa la segunda cifra: "))
  c=a+b
  print("El resultado eses: ",c,)
elif opcion == "B" or opcion=='b':
  print("Has elegido la opción de RESTAR, por lo cual debes ingresar dos cifras")
  d=float(input("Ingresa la primera cifra: "))
  e=float(input("Ingresa la segunda cifra: "))
  f=d-e
  print("El resultado es: ",f,)
elif opcion == "C" or opcion=='c' :
  print("Has elegido la opción de MULTIPLICAR, por lo cual debes ingresar dos cifras")
  g=float(input("Ingresa la primera cifra: "))
  h=float(input("Ingresa la segunda cifra: "))
  i=g*h
  print("El resultado es: ",i,)
elif opcion == "D" or opcion=='d':
  print("Has elegido la opción de DIVIDIR, por lo cual debes ingresar dos cifras")
  g=float(input("Ingresa la Numerador: ")) 
  h=float(input("Ingresa el Denominador: "))
  while h==0:
    h=float(input("Vuelva a ingresar un Denominador diferente de cero: "))
  i=g/h
  print("El resultado es: ",i,)
else:
  print("No has elegido ninguna de las opciones anteriores")
```

```
print("Ejercicio_6") 

print("En esta parte se determina el rango al que pertenece dependiendo sus")
print("ingresos mnesuales y cuanto es el porcentaje de impuestos a descontar.")
s=float(input("Para ello tendra que ingresar su salario en bruto: "))
if  s<10000:
  print("Usted entra en el primer Rango que es de menos de 10000 MXN")
  print("Porcentaje de impuestos para el rango: 5%")
  a=s*.05
  b=s-a
  print("Su salario es: ",b,)
elif 10000<=s<15000:
  print("Usted entra en el segundo Rango que es entre 10000 MXN y 15000 MXN")
  print("Porcentaje de impuestos para el rango: 10%")
  c=s*.10
  d=s-c
  print("Su salario es: ",d,)
elif 15000<=s<20000:
  print("Usted entra en el tercer Rango que es entre 15000 MXN y 20000 MXN")
  print("Porcentaje de impuestos para el rango: 15%")
  e=s*.15
  f=s-e
  print("Su salario es: ",f,)
elif s>20000:
  print("Usted entra en el cuarto Rango que en más de 20000 MXN")
  print("Porcentaje de impuestos para el rango: 18%")
  g=s*.15
  h=s-g
  print("Su salario es: ",h,)
else:
  print("Fin del Programa.")
```

```
print("Ejericico_9")

def factorial():
  f=1
  x=int(input("Ingrese una cifra entera para determinar su factorial: "))
  for i in range (1,x+1):
    f=f*i
  print("El resultado factorial es:", f,)
factorial()
```

```
print("En este paso podremos calcular el area de un circulo y el volumen de un cilidro")
pi = 3.1416
def volumen_cilindro(r, h):
    volumen = pi * r ** 2 * h
    print("El volumen  del cilindro es: ",volumen)  
def area_circulo(r):
    area = pi * r **2 
    print("El area del circulo es: ",area)
r=float(input("Ingrese el radio: "))
h=float(input("Ingrese la altura: "))
volumen_cilindro(r, h)
area_circulo(r)
```
